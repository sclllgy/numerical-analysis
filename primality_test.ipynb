### Eratosthenesの篩によって$n$以下の素数を出力するプログラム

```python
# ライブラリのインポート
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Eratosthenesの篩
def eratosthenes_sieve(n):
    p = [1]*(n+1)  # 0からnまでの配列を用意し，1で初期化する，1は素数であるとする．
    for k in range(2,int(np.sqrt(n))+1):
        if p[k]==1:
            for j in range(k,n):
                if k*j <= n:
                    p[k*j] = 0

    return p
```

```python
# n以下の素数を列挙
n = 100
p = eratosthenes_sieve(n)
for k in range(2,n):
    if p[k] == 1:
        print(k)
```

```python
# n以下の素数の個数を求める関数
def pn(n):
    num_p = []
    p = eratosthenes_sieve(n)  
    for k in range(2, n):
        num_p.append(sum(p[2:k+1]))

    return num_p
```

```python
n = 1000
x = np.linspace(2, n, n-2)  # 横軸の値を設定
pi = x / np.log(x)  # 素数定理の式
y = pn(n) * (np.log(x) / x)
```

```python
# matplotlibのレイアウト設定
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams["font.size"] = 16
```

```python
fig = plt.figure(figsize=(6,6))
ax1 = fig.add_subplot(2,1,1)
ax1.plot(pn(n), linewidth=2.0, label='$pn(n)$')
ax1.plot(x, pi, linewidth=2.0, linestyle='dashed', label='$\dfrac{n}{\log(n)}$')
ax1.set_title('Prime number theorem', fontsize=18)
ax1.set_ylabel('The number of primes', fontsize=16)
ax1.legend(loc='upper left', fontsize=16)
ax1.set_xlim(0, n)
ax1.grid(True)

ax2 = fig.add_subplot(2,1,2)
ax2.plot(x, y, linewidth=2.0)
ax2.set_xlabel('$n$', fontsize=16)
ax2.set_ylabel('$pn(n) \dfrac{\log(n)}{n}$', fontsize=16)
ax2.set_xlim(0, n)
ax2.grid(True)

plt.show()
```
