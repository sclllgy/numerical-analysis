### ニュートン法を用いて次の方程式の解を求めるプログラム
### $$ 1 = 3 \cdot 2^{-x} $$

```python
# ライブラリのインポート
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 関数func_fの定義
def func_f(x):
    return 3 * 2**(-x) - 1
```

```python
# func_fの導関数
def func_df(x):
    return -3 * 2**(-x) * np.log(2)
```

```python
# matplotlibのレイアウト設定
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams["font.size"] = 16
```

```python
# グラフの描画
x = np.linspace(-3, 3, 1000)
y = func_f(x)

fig = plt.figure(1)
ax = fig.add_subplot(1,1,1)
ax.plot(x, y, linewidth=2.0, label='$f(x)= 3 \cdot 2^{-x} - 1$')
ax.axhline(0.0, color='black', linewidth=1.0)
ax.set_xlabel('$x$', fontsize=16)
ax.set_ylabel('$f(x)$', fontsize=16)
ax.legend(loc='upper left', fontsize=16)
ax.grid(True)

plt.show()
```

```python
# ニュートン法
def newton(func_f, func_df, x0, esp=1e-4, max_loop=100):
    """
    param func_f:   対象となる関数
    param func_df:  func_fの導関数
    x0:             初期値
    param esp:      許容誤差
    param max_loop: 最大反復回数
    """
    count = 0  # 反復回数
    x = x0
    print("{:3d}: x = {:.15f}".format(count, x))

    while(True):
        # xの値を更新
        x -= func_f(x)/func_df(x)

        # 結果を表示
        count += 1  # 反復回数を数える
        print("{:3d}: x = {:.15f}".format(count, x))

        #「許容誤差がesp以下」または「反復回数がmax_loop以上」ならば終了
        if (abs(func_f(x)) <= esp) or (max_loop <= count):
            break

    # 最終的に得られた解
    print("x = {:.15f}".format(x))
```

```python
# ニュートン法による解
newton(func_f, func_df, 1.0)
```
