### 二分法を用いて次の方程式の解を求めるプログラム
### $$ 1 = 3 \cdot 2^{-x} $$

```python
# ライブラリのインポート
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 関数func_fの定義
def func_f(x):
    return 3 * 2**(-x) - 1
```

```python
# func_fの導関数
def func_df(x):
    return -3 * 2**(-x) * np.log(2)
```

```python
# matplotlibのレイアウト設定
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams["font.size"] = 16
```

```python
# グラフの描画
x = np.linspace(-3, 3, 1000)
y = func_f(x)

fig = plt.figure(1)
ax = fig.add_subplot(1,1,1)
ax.plot(x, y, linewidth=2.0, label='$f(x)= 3 \cdot 2^{-x} - 1$')
ax.axhline(0.0, color='black', linewidth=1.0)
ax.set_xlabel('$x$', fontsize=16)
ax.set_ylabel('$f(x)$', fontsize=16)
ax.legend(loc='upper left', fontsize=16)
ax.grid(True)

plt.show()
```

```python
# 二分法
def bisection(func_f, x_min, x_max, esp=1e-4, max_loop=100):
    '''
    param func_f:   対象となる関数
    param x_min:    探索区間の最小値
    param x_max:    探索区間の最大値
    param esp:      許容誤差
    param max_loop: 最大反復回数
    '''

    # 中間値の定理の条件を満たす場合
    if func_f(x_min)*func_f(x_max) < 0.0:
        count = 0  # 反復回数
        print("{:3d}: {:.15f} <= x <= {:.15f}".format(count, x_min, x_max))

        while(True):
            # 中間値の更新
            x_mid = (x_max + x_min)/2.0

            # 探索区間を更新
            if func_f(x_mid)*func_f(x_max) > 0.0:  # 中間と右端の値の符号が同じの時
                x_max = x_mid
            else:  # 中間と左端の値の符号が同じの時
                x_min = x_mid

            # 結果を表示
            count += 1  # 反復回数を数える
            print("{:3d}: {:.15f} <= x <= {:.15f}".format(count, x_min, x_max))

            #「許容誤差がesp以下」または「反復回数がmax_loop以上」ならば終了
            if (x_max-x_min <= esp) or (max_loop <= count):
                break

        # 最終的に得られた解
        print("x = {:.15f}".format(x_mid))

        return x_mid


    # 中間値の定理の条件を満たさない場合
    else:
        print("esp: The solution does not exist for closed ranges [{:.1f}, {:.1f}].".format(x_min, x_max))
```

```python
# 区間[1.0, 2.0]における解
bisection(func_f, 1.0, 2.0)
```
